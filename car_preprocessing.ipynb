```python
import pandas as pd
import numpy as np
```

# 수도권역

```python
dataset_seoul = pd.read_excel('car_seoul_2010.xlsx')
```

```python
dataset_seoul
```

```python
imsi_day = pd.DataFrame(dataset_seoul['일자'].dropna(axis = 0))
imsi_day = imsi_day.reset_index()
del imsi_day['index']
dataset_seoul = dataset_seoul[(dataset_seoul.입출구 == '합계')]
dataset_seoul = dataset_seoul.reset_index()
del dataset_seoul['index']
```

```python
del dataset_seoul['일자']
del dataset_seoul['입출구']
```

```python
dataset_seoul = pd.concat([imsi_day, dataset_seoul], axis = 1)
```

```python
day_time = []

for i in range(len(dataset_seoul)):
    for j in range(1,25):
        day_time.append(dataset_seoul.iloc[i][0] +'.' + dataset_seoul.columns[j])
```

```python
car_time = []

for i in range(len(dataset_seoul)):
    for j in range(1,25):
        car_time.append(dataset_seoul.iloc[i][j])
```

```python
dataset_seoul = pd.DataFrame({'시간' : day_time, '서울 교통량' : car_time })
```

```python
dataset_seoul = dataset_seoul.set_index('시간')
```

```python
dataset_seoul
```
