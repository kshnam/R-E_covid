# 교통량과 코로나

- 교통량 데이터는 2010년 이후로 존재하고, 코로나 데이터는 2020년 이후로 존재합니다.
- 표를 만들 수 있는 방법은 여러가지가 존재합니다.

1. 전국 교통량(평균) + 전국 코로나
2. 전국 교통량(따로) + 전국 코로나
3. 등등

### 1. 전국 교통량(평균) + 전국 코로나

```python
car_2020['합계'] = car_2020.sum(axis = 1)
```

```python
pd.DataFrame(car_2020['합계'])
```

```python
# 평균을 내야함
imsi_car_2020_daily = []

for i in range(int(len(car_2020)/24)):
    imsi_car = car_2020.iloc[24*i:24*(i+1)]
    imsi_car_sum = imsi_car.sum()
    imsi_car_2020_daily.append(imsi_car_sum)
    del imsi_car
    del imsi_car_sum
```

```python
''' 2020년도는 현재 교통량이 2020년 10월 25일까지밖에 없음. 9월'''
'''2019년도'''
imsi_day = pd.date_range(start = '2020-01-01', end = '2020-10-25')
```

```python
# 2020년도 각 일별 전국 교통량

a = pd.DataFrame(imsi_car_2020_daily)
b = pd.DataFrame(imsi_day)
c = pd.concat([a,b],axis = 1)
car_2020_daily = c.set_index(0)
del car_2020_daily.index.name
del imsi_day
```

```python
car_2020_daily
```

2019년 교통량도 위와같이 평균을 만들 수 있음

```python
car_2019['합계'] = car_2019.sum(axis = 1)

imsi_car_2019_daily = []
imsi_day = pd.date_range(start = '2019-01-01', end = '2019-12-31')
for i in range(int(len(car_2019)/24)):
    imsi_car = car_2019.iloc[24*i:24*(i+1)]
    imsi_car_sum = imsi_car.sum()
    imsi_car_2019_daily.append(imsi_car_sum)
    del imsi_car
    del imsi_car_sum

a = pd.DataFrame(imsi_car_2019_daily)
b = pd.DataFrame(imsi_day)
c = pd.concat([a,b],axis = 1)
car_2019_daily = c.set_index(0)
del car_2019_daily.index.name
del imsi_day
```

```python
car_2019_daily
```

코로나 데이터도 2020년 1월 1일부터 2020년 10월 25까지 표로 만들어주고, int로 변환합니다.

```python
# int 변환
corona_2020 = corona_2020.astype(int)

# 2020-01-01 ~ 2020-01-20까지 생성
imsi_corona_day = pd.DataFrame(['2020-01-01', '2020-01-02', '2020-01-03', '2020-01-04',
               '2020-01-05', '2020-01-06', '2020-01-07', '2020-01-08',
               '2020-01-09', '2020-01-10', '2020-01-11', '2020-01-12',
               '2020-01-13', '2020-01-14', '2020-01-15', '2020-01-16',
               '2020-01-17', '2020-01-18', '2020-01-19', '2020-01-20'])
imsi_zero = pd.DataFrame(np.zeros((len(imsi_corona_day),2),dtype = int))
imsi_zero = imsi_zero.rename(columns = {0 : '누적 확진자', 1 : '신규 확진자'})
imsi_concat = pd.concat([imsi_corona_day, imsi_zero],axis = 1)
imsi_concat = imsi_concat.set_index(0)
del imsi_concat.index.name
```

```python
# 자동차 정보량은 2020.10.25까지 있기 때문에 이후 코로나 데이터는 제거해준다.
# 2019년도도 비교를 위해선 2019.10.25까지로 바꿔준다.
imsi_corona = pd.concat([imsi_concat, corona_2020])
i_corona = imsi_corona.iloc[:-3]
car_2019_1025 = car_2019_daily.iloc[:-67]
```

```python
# index를 object에서 dt로 맞추기
imsi_day = pd.DataFrame(pd.date_range(start = '2020-01-01', end = '2020-10-25'))
corona = pd.concat([imsi_day,i_corona.reset_index()],axis = 1)
del corona['index']
corona = corona.set_index(0)
del corona.index.name
```
